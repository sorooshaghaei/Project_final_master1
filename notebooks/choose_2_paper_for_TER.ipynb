## 1) Paper A: `2_NeurIPS-2021-ttt-when-does-self-supervised-test-time-training-fail-or-thrive-Paper.pdf`

### What it solves

- Le TTT classique peut aggraver la performance sous fort shift de distribution parce que l'optimisation SSL (self-supervised learning) au test-time peut casser la représenattion utile pour la tâche principale.
- Proposition de TTT+, rendant la stabilité et l'effcacité sous shifts forts sans accès aux données sources.

### Main method idea

- Entraînement multi-tâche + SSL forte
- Offline feature summarization
- Test-Time Adaptation = SSL + Offline feature summarization

### Why it is good for TER

- Théorise quand le TTT thrive/fail
- Identification d'un échec concret du TTT
- Ajouts des gains nets sur plusieurs benchmarks (CIFAR-C, CIFAR10.1, VisDA‑C, toy moons) par rapport à BN-Adaptation, TENT, SHOT, TTT-R, ce qui illustre bien l'intérêt pratique du TTT avancé:

### Expected implementation blocks

1) Training stage
2) Features summarization stage (offline)
3) Test-Time adaptation Loop


## 2) Paper B: `4_ActMAD_Activation_Matching_to_Align_Distributions_for_Test-Time-Training.pdf.pdf`

### What it solves

- TTT plus générale : architecture-agnostic, task-agnostic (classification, detection, régression), online, source-free et data‑efficient.
- Part de l'idée de features alignment / critique le fait que la plupart des méthodes n'alignent que les canaux du dernier layer de l'encodeur / propose un alignement local et multi-couches des activations. 

### Main method idea

- ActMAD = Activation Matching to Align Distributions for Test-Time-Training

- Modéliser la distribution de chaque activation (C × H × W)
- Statistiques offline sur données clean
- Alignment online via backprop, sans SSL ni labels
- Plusieurs labels / update complet

### Why it is good for TER

- Méthode simple et plus générique (idéal pour discuter TTT au‑delà de la classification)
- Explication du cas online (flux, une seule passe) avec peu de données nécessaires pour converger.
- Démontration des résultats sur l'état de l'art ou quasi CIFAR‑10C/100C ImageNet‑C, et +15.4 points de mAP en détection sur KITTI‑Fog comparé au meilleur précédent, ce qui illustre bien la partie pratique.

### Expected implementation blocks

1) Offline
2) Online TTT loop
