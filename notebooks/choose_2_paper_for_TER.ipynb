# ActMAD Paper Selection Notes

These notes compare two TTT references that influenced the final project design.


## Compared Papers

| Paper | Main point | Use in this project |
|---|---|---|
| When Does Self-Supervised Test-Time Training Fail or Thrive? | TTT can help or hurt depending on the shift and adaptation objective | Motivation for reporting negative cases |
| ActMAD: Activation Matching to Align Distributions for Test-Time Training | Align source and target activation statistics during inference | Main adaptation idea implemented in the project |


## Selected Method

The final project uses an ActMAD-style method because it is simpler to implement, works without labels at test time, and fits naturally with saved source activation statistics.


## Implementation Scope

The project keeps the scope focused on CIFAR-10, CIFAR-10-C style corruptions, SimCLR pretraining, linear evaluation, and a baseline versus TTT comparison. The goal is a clear TER implementation, not a full reproduction of every experiment from the papers.
