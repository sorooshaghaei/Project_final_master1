## 1) Paper A: `2_NeurIPS-2021-ttt-when-does-self-supervised-test-time-training-fail-or-thrive-Paper.pdf`

### What it solves

- Classic TTT can make performance worse under strong distribution shift because SSL optimization at test time can damage the representation useful for the main task.
- TTT+ is proposed to improve stability and effectiveness under strong shifts without access to source data.

### Main method idea

- Multi-task training + strong SSL
- Offline feature summarization
- Test-Time Adaptation = SSL + Offline feature summarization

### Why it is good for TER

- Explains when TTT works or fails
- Identifies a concrete TTT failure
- Adds clear gains on several benchmarks (CIFAR-C, CIFAR10.1, VisDA-C, toy moons) compared with BN-Adaptation, TENT, SHOT, and TTT-R, showing the practical value of advanced TTT:

### Expected implementation blocks

1) Training stage
2) Features summarization stage (offline)
3) Test-Time adaptation Loop


## 2) Paper B: `4_ActMAD_Activation_Matching_to_Align_Distributions_for_Test-Time-Training.pdf.pdf`

### What it solves

- More general TTT: architecture-agnostic, task-agnostic (classification, detection, regression), online, source-free, and data-efficient.
- Starts from the idea of feature alignment, criticizes methods that only align the channels of the last encoder layer, and proposes local multi-layer activation alignment. 

### Main method idea

- ActMAD = Activation Matching to Align Distributions for Test-Time-Training

- Model the distribution of each activation (C x H x W)
- Offline statistics on clean data
- Online alignment via backprop, without SSL or labels
- Multiple labels / full update

### Why it is good for TER

- Simple and more general method, useful for discussing TTT beyond classification
- Explains the online case (stream, single pass) with little data needed to converge.
- Shows state-of-the-art or near state-of-the-art results on CIFAR-10C/100C and ImageNet-C, and +15.4 mAP in detection on KITTI-Fog compared with the previous best method, showing the practical side.

### Expected implementation blocks

1) Offline
2) Online TTT loop
