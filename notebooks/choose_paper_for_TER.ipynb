# TER Paper Selection Notes

These notes summarize the first paper comparison used to choose a feasible TER implementation direction.


## Initial Options

| Paper | Main idea | Implementation fit |
|---|---|---|
| Sun et al., 2020 | Test-Time Training with a self-supervised auxiliary task | Clear baseline, moderate scope |
| Self-supervised TTA direction | Test-time adaptation built around self-supervised representations | More ambitious, higher engineering risk |


## Decision

The project keeps the reliable TTT setting and combines it with self-supervised representation learning. The final implementation uses SimCLR pretraining on CIFAR-10, a frozen ResNet18 backbone for linear evaluation, and ActMAD-style activation matching during test-time adaptation.


## Rationale

This direction is realistic for the TER scope because it keeps the method simple, supports a clean baseline comparison, and still studies the main question: whether unlabeled target batches can help under image corruptions.
