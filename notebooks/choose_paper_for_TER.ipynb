Goal: choose one paper to implement for TER between:

This notebook is focused on **implementation feasibility + research value**.


## 1) Paper A: `1_sun20.pdf` (ICML 2020)

### What it solves
Improve robustness when train/test distributions differ by adapting at test time using unlabeled target samples.

### Main method idea
- Keep the main classification task.
- Add a self-supervised auxiliary task (rotation prediction in the paper).
- At test time, do a few gradient updates using the auxiliary loss, then predict labels.

### Why it is good for TER
- Clear algorithm and easy-to-defend baseline.
- Moderate implementation complexity.
- Strong starting point for later extensions.

### Expected implementation blocks
1. Backbone + classifier + rotation head.
2. Test-time adaptation loop (1 to 10 steps).
3. Corruption/domain-shift evaluation protocol.


## 2) Paper B: `3_When_test_time_Adaptation.pdf` (under review, ICLR 2026)

### What it solves
Study test-time adaptation for models trained with self-supervised learning, without depending on source-supervised pretraining.

### Main method idea
- Introduce self-supervised TTA protocol.
- Use a collaborative framework between SSL and TTA models.
- Combine contrastive learning, knowledge distillation, and mutual learning.

### Why it is interesting for TER
- Higher novelty and stronger SSL+TTA focus.
- Better if your TER goal is to push beyond baseline methods.

### Main risk
- Significantly more engineering effort and hyperparameter complexity than Paper A.
- Harder to finish cleanly if timeline is short.


## 3) Side-by-Side (Implementation View)

| Criterion | Paper A (Sun20) | Paper B (When-TTA-SSL) |
|---|---|---|
| Implementation effort | Medium | High |
| Reproducibility confidence | High | Medium |
| Risk for TER deadline | Low | Medium/High |
| Research novelty | Medium | High |
| SSL depth | Medium | High |
| Best use case | Reliable baseline + extension | Ambitious primary contribution |


## 4) Recommended Strategy

If you need reliability with good final quality:
1. Implement **Paper A** first as your strong baseline.
2. Add one extension inspired by **Paper B** (prototype classifier or mutual learning).

If your supervisor asks for maximum novelty and you have enough time/compute:
- Choose **Paper B** directly as main implementation target.


```python
# Simple scoring helper (edit weights and scores)
weights = {
    'deadline_safety': 0.35,
    'novelty': 0.25,
    'implementation_feasibility': 0.25,
    'ssl_depth': 0.15,
}

sun20 = {
    'deadline_safety': 9,
    'novelty': 6,
    'implementation_feasibility': 8,
    'ssl_depth': 6,
}

when_tta_ssl = {
    'deadline_safety': 6,
    'novelty': 9,
    'implementation_feasibility': 5,
    'ssl_depth': 9,
}

def weighted_score(scores, w):
    return sum(scores[k] * w[k] for k in w)

s1 = weighted_score(sun20, weights)
s2 = weighted_score(when_tta_ssl, weights)
print('Sun20 score:', round(s1, 3))
print('When-TTA-SSL score:', round(s2, 3))
print('Recommended:', 'Sun20' if s1 >= s2 else 'When-TTA-SSL')
```
