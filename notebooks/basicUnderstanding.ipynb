# Concept Notes

These notes summarize the main concepts used in the project.


## Distribution Shift

A classifier is usually trained on one data distribution and evaluated on another. On CIFAR-10-C, the test images can include noise, blur, fog, brightness changes, compression, or other corruptions. These shifts can reduce accuracy because the test images no longer match the clean training images.


## Self-Supervised Representation Learning

Self-supervised learning trains a model without class labels by creating a learning signal from the data itself. In this project, SimCLR creates two augmented views of the same CIFAR-10 image and trains the backbone to make their representations close while separating different images.


## Linear Evaluation

After SimCLR pretraining, the backbone is frozen. A linear classifier is trained on top of the frozen features. This gives a simple measure of representation quality without changing the backbone.


## Test-Time Training

Test-Time Training adapts the model on unlabeled test batches before prediction. The project uses ActMAD-style activation matching: source activation statistics are saved before evaluation, then the target batch is adapted by matching its internal statistics to the source statistics.


## Project Pipeline

1. Pretrain a ResNet18 backbone with SimCLR on CIFAR-10.
2. Train a linear classifier on frozen backbone features.
3. Save source activation statistics.
4. Evaluate clean baseline predictions and ActMAD-style TTT on corrupted images.
5. Compare accuracy, gains, and cases where TTT is unstable.
